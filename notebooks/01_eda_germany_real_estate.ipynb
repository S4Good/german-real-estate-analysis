```python
import numpy as np
import matplotlib.pyplot as plt

# Only take log of prices > 0
df["log_price"] = np.log10(df.loc[df["Price"] > 0, "Price"])

df["log_price"].dropna().plot(kind="hist", bins=50)
plt.title("Log Distribution of Property Prices (Germany)")
plt.xlabel("log10(Price)")
plt.ylabel("Count")
plt.show()
```

```python
df_clean = df[(df["Price"] > 10000) & (df["Price"] < 3000000) &
              (df["Living_space"] > 10) & (df["Living_space"] < 500)]

print("Original:", df.shape)
print("Cleaned :", df_clean.shape)
```

```python
df_clean = df[(df["Price"] > 10000) & (df["Price"] < 3000000) &
              (df["Living_space"] > 10) & (df["Living_space"] < 500)]

print("Original:", df.shape)
print("Cleaned :", df_clean.shape)
```

```python
top_cities = (
    df_clean.groupby("City")["Price"]
    .median()
    .sort_values(ascending=False)
    .head(10)
)

top_cities
```

```python
import matplotlib.pyplot as plt

top_cities.sort_values().plot(kind="barh")
plt.title("Top 10 Most Expensive Cities (Median Price)")
plt.xlabel("Median Price (€)")
plt.ylabel("City")
plt.show()
```

```python
import os
os.makedirs("../results/figures", exist_ok=True)

ax = top_cities.sort_values().plot(kind="barh")
plt.title("Top 10 Most Expensive Cities (Median Price)")
plt.xlabel("Median Price (€)")
plt.ylabel("City")
plt.tight_layout()
plt.savefig("../results/figures/top_10_cities_median_price.png", dpi=200)
plt.show()
```

```python
df_clean = df[(df["Price"] > 10000) & (df["Price"] < 3000000) &
              (df["Living_space"] > 10) & (df["Living_space"] < 500)].copy()

print("df_clean:", df_clean.shape)
```

```python
df_clean.loc[:, "price_per_m2"] = df_clean["Price"] / df_clean["Living_space"]
df_clean["price_per_m2"].describe()
```

```python
df_clean2 = df_clean[(df_clean["price_per_m2"] > 500) & (df_clean["price_per_m2"] < 20000)].copy()

print("Before:", df_clean.shape)
print("After :", df_clean2.shape)
```

```python
import matplotlib.pyplot as plt

df_clean2["price_per_m2"].plot(kind="hist", bins=60)
plt.title("Distribution of Price per m² (Germany)")
plt.xlabel("Price per m² (€)")
plt.ylabel("Count")
plt.show()
```

```python
state_prices = (
    df_clean2.groupby("State")["price_per_m2"]
    .median()
    .sort_values(ascending=False)
)

state_prices.head(10)
```

```python
import matplotlib.pyplot as plt

state_prices.head(10).sort_values().plot(kind="barh")
plt.title("Top 10 States by Median Price per m²")
plt.xlabel("Median €/m²")
plt.ylabel("State")
plt.show()
```

```python
import os
os.makedirs("../results/figures", exist_ok=True)

ax = state_prices.head(10).sort_values().plot(kind="barh")
plt.title("Top 10 States by Median Price per m²")
plt.xlabel("Median €/m²")
plt.ylabel("State")
plt.tight_layout()
plt.savefig("../results/figures/top_10_states_price_per_m2.png", dpi=200)
plt.show()
```

## Key Findings (Summary)

- The dataset contains **10,552 listings** across Germany.
- Many columns have missing values, especially energy-related data.
- Property prices are strongly right-skewed (few very expensive listings), so I used a **log scale** to visualize the distribution.
- After removing extreme outliers, the cleaned dataset contains **10,055 listings**.
- Median price varies strongly by location:
  - **Top expensive cities** include premium districts (e.g., Berlin districts and high-demand Bavarian regions).
- For better comparison, I calculated **Price per m² (€/m²)**.
  - After removing unrealistic €/m² outliers, **9,685 listings** remained.
  - **Hamburg and Berlin** show the highest median €/m² among the top states.
